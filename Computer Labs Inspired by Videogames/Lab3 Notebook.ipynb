## Simulating Sample Statistics

```r
#Create some population
N = 1000
data_pop = rbinom(n=N, size = 1, prob=0.3) #Random binomial with 1 choice, aka bernoulli distribution

#Grab a sample of the population
n = 300
data_sample = sample(data_pop, size = n, replace = FALSE) #Sample randomly from the dataset
#Alternately do the sampling manually with row indices
#ind_sample = sample.int(n=N, size=n) #Randomly samples 300#'s from 1000#'s. Use this as our index.
#data_sample = data_pop[ind_sample] #Select the random sample with this index
data_sample
```

```r
# A simple confidence interval
width = 2*sd(data_sample)/sqrt(n)
confint_simple = c(mean(data_sample)-width, mean(data_sample)+width) #This is a choose function
confint_simple
```

```r
# A correction confidence interval (corrected for finite sample population)
m = mean(data_sample)
width = 1.96*sqrt(m*(1-m)*(N-n)/((n-1)*N))
confint_simple = c(m-width, m+width)
confint_simple
```

For the above code, to get an interval estimate for the fraction, we follow the derivation in the lecture slides, the interval estimate is then given by,
$$ \left(\bar x - 1.96 \sqrt{\frac{\bar x (1 - \bar x)}{n-1} \frac{N - n}{N}}, \bar x + 1.96 \sqrt{\frac{\bar x (1 - \bar x)}{n-1} \frac{N - n}{N}} \right), $$
